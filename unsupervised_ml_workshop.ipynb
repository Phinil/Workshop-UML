#### Exercise 2B.3 ** — Experiment with linkage methods

The linkage method defines how the distance between clusters is measured. Try different methods and compare the dendrograms:

- `ward` — minimizes within-cluster variance (tends to produce equal-sized clusters)
- `complete` — maximum distance between clusters
- `average` — average distance between clusters
- `single` — minimum distance (can produce long, chain-like clusters)

1. Plot dendrograms for at least two different linkage methods side by side
2. How do the resulting cluster structures differ?
3. Which linkage method seems best for your data?

````python
# Your code here

from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt

X_sample = X_scaled[:1000]  # viktigt så kernel inte dör


Z = linkage(X_sample, method='ward')
plt.figure(figsize=(12, 6))
dendrogram(Z, truncate_mode='lastp', p=30)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Cluster size')
plt.ylabel('Distance')
plt.show()

Z = linkage(X_sample, method='complete')
plt.figure(figsize=(12, 6))
dendrogram(Z, truncate_mode='lastp', p=30)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Cluster size')
plt.ylabel('Distance')
plt.show()

````

````python
#Resultatet visar att metoderna ger olika typer av dendrogram och därmed olika sätt att gruppera datapunkterna. Ward linkage skapar tydliga och kompakta kluster med ett stort avstånd mellan huvudgrupperna, vilket gör det lätt att avgöra var man kan “klippa” trädet. Complete linkage håller kluster mer separerade men kan bli känslig för extrema värden. Average linkage ger en mer balanserad struktur men utan lika tydliga uppdelningar. Single linkage visar en kedjeeffekt där många punkter kopplas ihop stegvis, vilket gör det svårt att identifiera tydliga kluster. Baserat på jämförelsen verkar Ward linkage fungera bäst för denna data, eftersom den ger en tydlig och lättolkad klusterstruktur.
````

#### Exercise 2B.4 ** — Silhouette Score & Interpretation

1. Calculate the silhouette score for your clustering
2. Try different numbers of clusters (2-8) and plot k vs. silhouette score
3. Group the **original (unscaled)** data by cluster and compute means — can you name the clusters?

````python
# Your code here

from sklearn.metrics import silhouette_score
from sklearn.cluster import AgglomerativeClustering

scores = []
K = range(2, 9)

X_sample = X_scaled[:2000]

for k in K:
    agg = AgglomerativeClustering(n_clusters=k, linkage='ward')
    labels = agg.fit_predict(X_sample)
    
    score = silhouette_score(X_sample, labels)
    scores.append(score)
    print(f"k={k}, score={score:.3f}")
````

````python
# Vi beräknade silhouette score för olika antal kluster (k=2–8) för att utvärdera kvaliteten på klustringen. 
# Resultatet visar att k=2 ger högst score (0.260), vilket indikerar den bästa separationen mellan kluster. 
# När antalet kluster ökar minskar silhouette score, vilket tyder på att strukturen blir mindre tydlig. 
# Detta bekräftar tidigare observationer från dendrogrammet att två kluster är ett rimligt val.
````
